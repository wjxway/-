### Python 有迭代器，能解决内存问题

```python
# Python
from itertools import permutations

# 生成器函数写法

def find_sth():
    for i in permutations(range(1, 5)):
        if i[0] == 1:
            yield i

# 生成器写法
            
find_sth1=(i for i in permutations(range(1, 5)) if i[0]==1)

for j in find_sth():
    print(j)

print()
# 隐式迭代器，写起来方便

for j in find_sth1:
    print(j)
```

### 结论

我们可以在 notebook/Jupyter 中运行 Python，`from itertools import permutations`

生成的结果是 wl 表达式，我们可以进行后续操作

利用 Python 的迭代器来解决 wl 经常爆内存的问题

也就是所谓的流

### 接下来是 Mathematica 的模式匹配

我们这里就以 `Cases`/`MatchQ` 为例

直接把整个 `MatchQ` 连同 `Pattern` 写成纯函数，并把它赋给一个 `foo`

之后就能畅快地使用 Python 的迭代器，+ Mathematica 的模式匹配了

> 原本需要写
>
> foo = Eval(wlexpr('str')) 来定义函数
>
> Eval(foo(para)) 来计算
>
> 但是利用高阶过程改写函数就很自然

https://reference.wolfram.com/language/WolframClientForPython/

网址的写法不算太方便

```python
from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr

from numpy import array

session = WolframLanguageSession()

Eval = session.evaluate # 起别名

def foo(bar):
    temp = Eval(wlexpr('Cases[{___,1,1,___}]'))
    return Eval(temp(bar))

foo([[1,1,2],[3,2,1],[0,1,1,3]])
```

```python
from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr, Global
from itertools import permutations
session = WolframLanguageSession()
Eval = session.evaluate
def foo1(bar):
    temp = Eval(wlexpr('MatchQ[{__,1,2,__}]'))
    return Eval(temp(bar))
# 可以用生成器
def temp():
    for i in permutations(range(14)):
        if foo1(i):
            print(i)
```

### 也可以写成 lambda

```python
from numpy import array
foo2 = lambda z:Eval(
    Eval(
    wlexpr('EvenQ')
)(z)
)

[foo2(i) for i in array([1,2,3,4],dtype='int8')]
```

### 注意事项

`wl` 可以调用函数，但是只能以函数名+参数的方法。

比如 `wl.Times(wl.Plus(1,2),3)` 可以

`wl.Plus(1,2)*3` 不行

```python
Eval(wl.Times(wl.Plus(1,2),3))
```

```python
Eval(wl.Plus(1,2)*3)
# TypeError: unsupported operand type(s) for *: 'WLFunction' and 'int'
```

所以需要<del>手写 FullForm</del>

把操作符定义成函数调用

```python
foo3=Eval(wlexpr('#3(#+#2)&'))
Eval(foo3(1,2,4))
```

### 再举个例子，写一个简易求解器

#### FullForm 写法

```python
x = Global.x
Eval(wl.Solve(wl.Equal( wl.Plus(x,2),0),x))
```

#### 函数调用写法

```python
def solve(n):
    return Eval(Eval(wlexpr("Values@Solve[x+#==0,x]&"))(n))[0][0]

# SQL
```

```python
solve(10)
```

```python
solve(5)
```

## 拓展例子
在 Mathematica 里，可以直接用 1-图片 来反色

```python
def rev(n):
    return Eval(Eval(wlexpr("1-#&"))(n))
```
